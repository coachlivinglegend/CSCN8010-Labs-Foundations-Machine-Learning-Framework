## Cleaning the Data

### Remove Capped Data

```python
housing = housing[housing['median_house_value'] < 500000]
```

### Remove Missing Data

```python
missing_data = housing.isnull().sum()
missing_data
```

We have 162 rows where total_bedrooms are missing. For this example, we will just remove these rows, in other instances, we can replace the null values, with the mean of the other values or sd or md or whatever.

```python
housing = housing[housing['total_bedrooms'].notna()]
```
