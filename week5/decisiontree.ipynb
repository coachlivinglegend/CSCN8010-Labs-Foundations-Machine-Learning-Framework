```python
# load diabetes dataset
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target

# filter for the BMI feature
X_BMI = X[:,2].reshape(-1,1)
```

```python
from sklearn import tree

clf = tree.DecisionTreeRegressor(max_depth=2)
clf = clf.fit(X_BMI, y)
```

```python
# plot the data as a scatter plot, and the decision tree as a line
import matplotlib.pyplot as plt
plt.scatter(X_BMI, y, c="orange")

# create a range of values for the x-axis
import numpy as np
tree_x = np.linspace(min(X_BMI), max(X_BMI), 1000).reshape(-1,1)

plt.plot(tree_x, clf.predict(tree_x), c="black")

# add labels
plt.xlabel("BMI")
plt.ylabel("Disease Progression")
plt.title("Decision Tree Regression")

plt.show()
```
